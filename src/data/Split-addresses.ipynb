```python
import json
import re
```

```python
f = open('addresses.json')
data = json.load(f)
```

```python
splitted_addresses = []
```

```python
for project in data:
    address = project["adres"]
    splitted = re.split('\s',address)
    
    if len(splitted) != 4:
        continue
        
    address_info = {}
    address_info["streetName"] = splitted[0]
    address_info["houseNumber"] = splitted[1].replace(",","")
    address_info["postalCode"] = splitted[2]
    address_info["city"] = splitted[3]
    
    address_info["projectID"] = project["project"]
    splitted_addresses.append(address_info)
```

```python
splitted_addresses
```

```python
with open("splitted-addresses.json","w") as outfile:
    json.dump(splitted_addresses, outfile)
```
