```python
import json
import re
```

```python
f = open('addresses.json')
data = json.load(f)
```

```python
splitted_addresses = []
```

```python
for project in data:
    address = project["adres"]
    #splitted = re.split('\s',address)
    splitted = re.findall("([a-zA-Z]+(?:\s[a-zA-Z])*){1}\s?(\d*),\s(\d+){1}\s(\w+){1}", address)[0]
    
    #if len(splitted) != 4:
        #continue
        
    address_info = {}
    address_info["streetName"] = splitted[0]
    address_info["streetNumber"] = splitted[1] if splitted[1] else project["naam"]
    address_info["postalCode"] = splitted[2]
    address_info["city"] = splitted[3]
    
    address_info["projectID"] = project["project"]
    splitted_addresses.append(address_info)
```

```python
splitted_addresses
```

```python
with open("splitted-addresses-incl-no-street-number.json","w") as outfile:
    json.dump(splitted_addresses, outfile)
```
